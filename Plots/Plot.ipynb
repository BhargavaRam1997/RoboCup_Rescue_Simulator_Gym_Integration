```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
```

```python
file = r'Plot.xlsx'
df = pd.read_excel(file)
y_ppo = list(df['PPO'])
y_ppo_error = list(df['Standard deviation'])
y_greedy = list(df['Greedy'])
y_dqn = list(df['HDQN'])
y_dqn_error = list(df['Standard deviation']) 
y_a2c = list(df['A2C'])
y_a2c_error = list(df['Standard deviation'])
```

```python
x = range(0,100*len(y_ppo),100)

plt.figure(figsize=(15,10))
y_ppo_neg_error = [y_ppo - y_ppo_error for y_ppo, y_ppo_error in zip(y_ppo, y_ppo_error)]
y_ppo_pos_error = [y_ppo + y_ppo_error for y_ppo, y_ppo_error in zip(y_ppo, y_ppo_error)]
y_dqn_neg_error = [y_dqn - y_dqn_error for y_dqn, y_dqn_error in zip(y_dqn, y_dqn_error)]
y_dqn_pos_error = [y_dqn + y_dqn_error for y_dqn, y_dqn_error in zip(y_dqn, y_dqn_error)]
y_a2c_neg_error = [y_a2c - y_a2c_error for y_a2c, y_a2c_error in zip(y_a2c, y_a2c_error)]
y_a2c_pos_error = [y_a2c + y_a2c_error for y_a2c, y_a2c_error in zip(y_a2c, y_a2c_error)]
y_max = [10]*len(y_ppo)
y_greedy = [8.57]*len(y_ppo)
## PPO

plt.plot(x, y_ppo, 'b-', label='PPO')
plt.fill_between(x, y_ppo_neg_error, y_ppo_pos_error , alpha=0.4)

## HDQN

# plt.plot(x, y_dqn, 'r-', label = 'HDQN')
# plt.fill_between(x, y_dqn_neg_error, y_dqn_pos_error, alpha=0.4)

# A2C

# plt.plot(x, y_a2c, 'p-', label='A2C')
# plt.fill_between(x, y_a2c_neg_error, y_a2c_pos_error, alpha=0.4)

# Max and greedy
plt.plot(x, y_max, 'g-', label = 'Maximum Reward')
plt.plot(x, y_greedy, 'y-', label = 'Greedy')
plt.xlabel("Number of episodes")
plt.ylabel("Cumulative Reward")
# plt.title("Big Map | Iterations = 5")
# plt.title("Agents ignoring commands until they reach their destination")
plt.legend(loc="upper left")
# plt.xticks(np.arange(11000,step=1000))
plt.ylim(bottom=7, top=12)
plt.show()
```
